# Chat Bot

```python
from langchain.document_loaders import PyPDFLoader
import os

directory = '/content/data'
documents = []

for filename in os.listdir(directory):
    if filename.endswith('.pdf'):
        file_path = os.path.join(directory, filename)
        loader = PyPDFLoader(file_path)
        documents.extend(loader.load())

print(f"Number of documents loaded: {len(documents)}")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

def split_docs(documents,chunk_size=500,chunk_overlap=20):
  text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
  docs = text_splitter.split_documents(documents)
  return docs

docs = split_docs(documents)
print(len(docs))
```

```python
import os
from pinecone import Pinecone, ServerlessSpec
from langchain.vectorstores import Pinecone as LangchainPinecone

# Set the Pinecone API key as an environment variable
os.environ["PINECONE_API_KEY"] = "eb1e40f5-68fc-443c-9e1b-78ac82a54e2b"

# Membuat instance Pinecone
pc = Pinecone()

# Mengecek apakah index sudah ada atau belum
index_name = "chatbot-ptik"
if index_name not in pc.list_indexes().names():
    pc.create_index(
        name=index_name,
        dimension=384,  # Sesuaikan dengan dimensi embeddings Anda
        metric='cosine',
        spec=ServerlessSpec(
            cloud='aws',
            region='us-east-1'  # Ganti dengan region yang sesuai
        )
    )

# Make sure your 'docs' variable is defined and contains your documents
# For example:
# from langchain.document_loaders import TextLoader
# loader = TextLoader('./my_document.txt')
# docs = loader.load()

# Menggunakan index yang sudah ada atau baru saja dibuat
index = LangchainPinecone.from_documents(docs, embeddings, index_name=index_name)
```

```python
def get_similiar_docs(query,k=1,score=False):
  if score:
    similar_docs = index.similarity_search_with_score(query,k=k)
  else:
    similar_docs = index.similarity_search(query,k=k)
  return similar_docs

query = "How can i register my krs"
similar_docs = get_similiar_docs(query)
similar_docs
```

```python
# prompt: sekarang saya mau apapun query dari user maka responsenya otomatis memakai bahasa user misal user masuk query bahasa inggris maka cari jawabannya kemudian translate jawaban ke inggris

!pip install googletrans==4.0.0-rc1

from googletrans import Translator
translator = Translator()

def get_response(query):
  similar_docs = get_similiar_docs(query)
  # Assuming the first similar document contains the relevant answer
  answer = similar_docs[0][0].page_content

  detected_lang = translator.detect(query).lang
  translated_answer = translator.translate(answer, dest=detected_lang).text

  return translated_answer

# Example usage
user_query = "How can I register my krs?"
response = get_response(user_query)
print(response)
```
